```python
# Factorial of a number

n = int(input("Enter a number: "))
fact = 1

while n > 0:
  fact *= n
  n -= 1

print("Factorial:", fact)
```

```python
# Fibonacci seriesusing while loop

n = int(input("Enter the number of terms: "))

a = 0
b = 1
i = 1
while i <= n:
  print(a, end=" ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
# Reverse a number
num = int(input("Enter a number: "))
rev = 0
while num > 0:
  digit = num % 10
  rev = rev * 10 + digit
  num //= 10

  print("Reverse:", rev)
```

```python
# Check Palindrome number/string
num = int(input("Enter a number: "))
temp = num
while num > 0:
    rev = rev * 10 + num % 10
    num //= 10
if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Is a palindrome
s = "madam"
i, j = 0, len(s) - 1
is_palindrome = True
while i<j:
  if s[i] != s[j]:
     is_palindrome = False
     break
  i += 1
  j -= 1
if is_palindrome:
  print("Yes")
else:
  print("No")
```

```python
# I s not a palindrome
s = "Hello"
i, j = 0, len(s) - 1
is_palindrome = True
while i<j:
  if s[i] != s[j]:
     is_palindrome = False
     break
  i += 1
  j -= 1
if is_palindrome:
  print("Yes")
else:
  print("No")
```
